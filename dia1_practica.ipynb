```python
# Crear lista de numero del 1 al 10
numeros = list(range(1, 11))
print(numeros)

# Filtrar pares con loop tradicional
pares_loop = []
for n in numeros:
    if n % 2 == 0:
        pares_loop.append(n)
print(f"Pares (loop): {pares_loop}")

# con list comprehension

pares_comp = [n for n in numeros if n % 2 == 0]
print(f"Pares (com): {pares_comp}")

# Transformar: cuadrados de impares
cuadrados_impares = [n**2 for n in numeros if n % 2 != 0]
print(f"Cuadrados de impares: {cuadrados_impares}")
```

```python
# Dataset simulados: informacion de estudiantes
estudiantes = {
    "Ana": {"edad": 25, "nota": 8.5, "ciudad": "Madrid"},
    "Luis": {"edad": 30, "nota": 7.0, "ciudad": "México"},
    "Carla": {"edad": 22, "nota": 9.2, "ciudad": "Buenos Aires"},
}

# Accedear a datos
print(f"Nota de Ana:{estudiantes['Ana']['nota']}")

# filtrar: estudiantes con nota > 8
aprobados_alto = {
    nombre: datos for nombre, datos in estudiantes.items() if (datos["nota"] > 8)
}
print(f" Aprovado con mas de 8: {aprobados_alto}")

# Agregar campo a calculado: Categoria
for nombre, datos in estudiantes.items():
    if datos["nota"] >= 9:
        datos["categoria"] = "Excelente"
    elif datos["nota"] >= 7:
        datos["categoria"] = "Bueno"
    else:
        datos["categoria"] = "Regular"
print(f"Estudiantes: {estudiantes}")        
```

```python
# Funciones

def calcular_estadisticas(numeros):
    """
    Recibe lista de numeros, retorna diccionario con 
    estadisticas.
    """
    if not numeros: # Manejo de caso vacio
        return None
    return {
        "cantidad": len(numeros),
        "suma": sum(numeros),
        "promedio": sum(numeros) / len(numeros),
        "maximo": max(numeros),
        "minimo": min(numeros)
    }
    
datos= [23,45,67,12,89,34]
stats = calcular_estadisticas(datos)
print(stats)    


# Aplicar a grupos de datos
grupos={
    "Grupo A":[85,90,78,92],
    "GRUPO B":[70,75,80,65],
    "GRUPO C":[95,98,92,96]
}

resultados = {nombre: calcular_estadisticas(notas)
              for nombre, notas in grupos.items()
              }

print(resultados)

```
